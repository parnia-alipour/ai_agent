```python
import gradio as gr

from langchain_openai import ChatOpenAI
```

```python
base_url="https://openrouter.ai/api/v1"
api_key=""
model_name="gpt-4o"
```

```python
llm=ChatOpenAI(
    base_url=base_url,
    api_key=api_key,
    model=model_name
)

def llm_g(model_name):
    return ChatOpenAI(
    base_url=base_url,
    api_key=api_key,
    model=model_name
)
def chatbot(user_input,history,model_name):
    llm=llm_g(model_name)
    message=[
        {"role":"system",
         "content":"Whatever language the person speaks, respond in that language."},
        {"role":"system",
        "content":"You are a professional psychologist who helps others and makes them well."}]

    for user_msg,bot_msg in history:
        message.append({"role":"user","content":user_msg})
        message.append({"role":"assistant","content":bot_msg})

    message.append({"role":"user","content":user_input})
    ai_message=llm.invoke(message)
    return ai_message.content
chatbot2=gr.ChatInterface(
    fn=lambda message,history:chatbot(message,history,model_name),
    title="Chatbot  with gpt",
    description="This chatbot is built with the goal of being a psychoanalyst and therapist for the user. You can actually talk to it via the API.",
    )
chatbot2.launch()
```
