```python
import gradio as gr
from accelerate.commands.config.update import description
from openai import OpenAI
from langchain_openai import ChatOpenAI
```

```python
base_url="https://openrouter.ai/api/v1"
api_key=""
model_name="gpt-4o_mini"
```

```python
def llm(model_name):
        return ChatOpenAI(api_key=api_key
               ,base_url=base_url
               ,model=model_name,)

def chatbot(history,user_inputs,model_name):
    llm_2=llm(model_name)
    message=[{
        "role":"system",
        "content":"Summarize any text the user provides and present it in a completely useful way."},{
            "role":"system",
            "content":"You are the one who summarizes every text very professionally. You use useful words. If the user asks you not to use difficult words, you don't use them. Even if the user doesn't insist on this, you provide the best text that you think is perfect in every way."

    }]

    for user_msg,bot_msg in history:
        message.append({"role":"user","content":user_msg})
        message.append({"role":"assistant","content":bot_msg})

    message.append({"role":"user","content":user_inputs})
    ai_message=llm_2.invoke(message)
    return ai_message.content
chatbot2=gr.ChatInterface(
fn=lambda  message,history:chatbot(history,message,model_name),
    title="A chatbot that writes summaries",
    description="Give any article or text to the chatbot and it will easily summarize it for you using its API."

)
chatbot2.launch()
```
