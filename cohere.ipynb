```python
import cohere
```

```python
co=cohere.ClientV2(api_key="your_api_key")

res=co.chat(
    model="command-r-plus-08-2024",
    messages=[{
        "role":"user",
        "content":"What do you think AI is?"

    }
    ]
)
print(res.message.content[0].text)
```

```python
co=cohere.ClientV2(api_key="your_api_key")
system_message="you response concisely, in about 20 words less"
res=co.chat(
    model="command-r-plus-08-2024",
    messages=[
        {"role":"system","content":system_message},
         {
        "role":"user",
        "content":"Is carrot juice good for your health?"
         }
    ],
)
print(res.message.content[0].text)
```
