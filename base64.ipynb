```python
import os.path

from openai import OpenAI
import base64
import os
```

```python
i="output.png"
with open (i,"rb") as f :
    encoding_string=base64.b64encode(f.read()).decode("utf-8")
    print("base64:",encoding_string[:800],"...")
    print("len base64:",len(encoding_string))
```

```python
file_name,file_ext=os.path.splitext(i)
output_file=f"{file_name}-base64{file_ext}"

with open (output_file,"wb") as f :
    f.write(base64.b64decode(encoding_string))

print(f"file:{output_file}")
```
