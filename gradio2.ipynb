```python
import gradio as gr
from openai import OpenAI
import os
import base64
from io import BytesIO
from langchain_openai import ChatOpenAI

```

```python
base_url = "https://api.avalai.ir/v1"
api_key = ""
model="gpt-4o"
```

```python
from io import BytesIO
import base64

llm = ChatOpenAI(
    base_url=base_url,
    api_key=api_key,
    model=model
)

def encode_image_to_base64(image):
    buffer = BytesIO()
    image.save(buffer, format="PNG")
    return base64.b64encode(buffer.getvalue()).decode("utf-8")

def analyze_image(image):
    try:
        image_base64 = encode_image_to_base64(image)
        image_ext = "png"
        message = [
            {"role": "system", "content": "you are a helpful persian assistant"},
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": "به فارسی بگو چی میبینی تو عکس"},
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": f"data:image/{image_ext};base64,{image_base64}",
                            "detail": "auto"
                        }
                    }
                ]
            }
        ]
        print("Function triggered")

        ai_message = llm.invoke(message)
        return ai_message.content

    except Exception as e:
        return str(e)

iface=gr.Interface(
    fn=analyze_image,
    inputs=gr.Image(type="pil"),
    outputs=gr.Textbox(label="Image description with artificial intelligence"),
    title="Image analysis with artificial intelligence",
    description="Provide an image so that artificial intelligence can give a description or explanation about it.",
)
iface.launch()
```
