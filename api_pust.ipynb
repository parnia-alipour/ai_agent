```python
import requests
```

```python
url="http://jsonplaceholder.typicode.com/posts"
```

```python
order={
    "customer":"parnia",
    "predict":"loptop",
    "quantity":2,
    "price":35000
}
```

```python
response=requests.post(url,json=order)
```

```python
if response.status_code==201:
    data=response.json()
    print("order har been registered")
    print("server response")
    print(data)

else:
    print("Error:",response.status_code)
```
