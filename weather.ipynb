```python
import requests
```

```python
url="https://api.open-meteo.com/v1/forecast?latitude=34.0917&longitude=49.6892&current=temperature_2m,relative_humidity_2m,wind_speed_10m"
```

```python
response=requests.get(url)
response.status_code
```

```python
data=response.json()
```

```python
print("Temperature:",data['current']['temperature_2m'])
print("Relative Humidity:",data['current']['relative_humidity_2m'])
print("Wind Speed:",data['current']['wind_speed_10m'])
```

```python
url2="https://postman-echo.com/post"
```

```python
response2=requests.post(url=url2,json=data)
response2.status_code
```
