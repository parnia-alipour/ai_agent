```python
import requests
```

```python
url="https://api.coingecko.com/api/v3/simple/price?ids=bitcoin&vs_currencies=usd,eur"
response=requests.get(url)
print(response.status_code)
```

```python
if response.status_code==200:
    data=response.json()
    usd_price=data['bitcoin']['usd']
    eur_price=data['bitcoin']['eur']
    usd_to_toman = 1600000
    irr_price = usd_price * usd_to_toman
    irr_price = float(irr_price)
    print("قیمت لحظه ای بیت کوین:")
    print(f"USD {usd_price} $")
    print(f"EUR {eur_price}")
    print(f"IRR {irr_price:,.0f}")
else:
    print("Error:",response.status_code)
```
